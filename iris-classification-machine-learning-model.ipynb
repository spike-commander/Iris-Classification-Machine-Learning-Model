# Second Machine Learning Model (Iris Detection)

## Library Imports (Numpy, Pandas)

```python
import pandas as pd
import numpy as np
```

## Data Prepration

**Species Convertion: setosa -> 1, virgininca -> 2, versicolor -> 3**

```python
data = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/iris.csv')
species_map = {
    'setosa': 1,
    'virginica': 2,
    'versicolor': 3
}
data['Species'] = data['Species'].map(species_map)
data
```

## Data Sepration (into X and Y )

```python
y = data['Species']

x = data.drop('Species', axis = 1)
```

```python
from sklearn.model_selection import train_test_split

x_test, x_train, y_test, y_train = train_test_split(x, y, test_size = 0.2, random_state = 100)

y_train
```

## Model Building

**Should not use the Linear Regression Model, It's just predicts contiunes numbers only, dosen't classifies**

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)

LinearRegression()
```

### Applying Model to make Prediction

```python
y_lr_train_pred = lr.predict(x_train).astype(int)
y_lr_test_pred = lr.predict(x_test).astype(int)

y_lr_train_pred
```

## Evaluate Model Performance (Linear REgression)

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)

print("lr_train_mse = ", lr_train_mse)
print("lr_train_r2 = ", lr_train_r2)
print("---------------------------------------")
print("lr_test_mse = ", lr_test_mse)
print("lr_test_r2 = ", lr_test_r2)
```

## Classification Model

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report

svm = SVC(kernel='linear', random_state=42)
svm.fit(x_train, y_train)
svm
```

### Applying Model To make Prediction

```python
svm_pred = svm.predict(x_test)
svm_pred
```

## Classification Report

```python
print(classification_report(y_test, svm_pred))
```

## Applying the Model to make Prediction on a real Data ( By User )

```python
def pred_model():
  print("You have chosen the 'Know Your Iris Type' Option!")
  try:
    sepal_len = float(input("Enter your Sepal Lenght: "))
    sepal_wid = float(input("Enter your Sepal Width: "))
    petal_len = float(input("Enter your Petal Lenght: "))
    petal_wid = float(input("Enter your Petal Width: "))

    # Create a DataFrame from the user input
    user_input = pd.DataFrame([[sepal_len, sepal_wid, petal_len, petal_wid]],
                              columns=['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width'])

    # Predict using the trained SVM model
    prediction = svm.predict(user_input)

    # Create a reverse map for species names
    reverse_species_map = {v: k for k, v in species_map.items()}
    predicted_species = reverse_species_map[prediction[0]]

    print(f"\nBased on the input, the Iris type is: {predicted_species}")
  except ValueError:
    print("Invalid input. Please enter numerical values for measurements.")
  except KeyError:
    print("Prediction out of expected range. The model might need more diverse training data.")
```

```python
while True:
  print("\n--- Iris Classification System ---\n")
  print("1. Know your Iris type")
  print("2. Exit")

  choice = input("Enter your choice (1 or 2): ")

  if choice == '1':
    pred_model()
  elif choice == '2':
    print("Exiting Iris Classification System. Goodbye!")
    break
  else:
    print("Invalid choice. Please enter 1 or 2.")
```
